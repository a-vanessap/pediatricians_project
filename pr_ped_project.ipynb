```python
import pandas as pd
df = pd.read_csv("ahrf2024_Feb2025.csv", low_memory=False)
```

```python
df.head()
```

```python
df_pr = df[df['st_name'] == 'Puerto Rico']
df_pr.head()
```

```python
pop_columns = [
    'popn_mal_lt5_20', 'popn_fem_lt5_20',
    'popn_mal_5_9_20', 'popn_fem_5_9_20',
    'popn_mal_10_14_20', 'popn_fem_10_14_20',
    'popn_mal_15_19_20', 'popn_fem_15_19_20'
]
df_pr.loc[:, 'kid_pop_total'] = df_pr[pop_columns].sum(axis=1)
```

```python
df_ped = df_pr[['cnty_name', 'md_nf_ped_gen_22', 'kid_pop_total']]
df_ped.head()
```

```python
median_kid_pop = df_ped['kid_pop_total'].median()
print(median_kid_pop)
```

```python
df_ped['peds_per_child'] = df_ped['md_nf_ped_gen_22'] / df_ped['kid_pop_total'] * 6278
df_ped.head()
```

```python
df_ped.to_csv('pr_ped_project.csv', index=False)
```
